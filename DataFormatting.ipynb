```python
#the ratio of women to men over the years

gendergrouping2017 = df2017.groupby('Gender').count()
gendergrouping2017= gendergrouping2017['num']

ratio2017 = gendergrouping2017['Male'] /gendergrouping2017['Female']

gendergrouping2018 = df2018.groupby('Gender').count()
gendergrouping2018 = gendergrouping2018['num']

ratio2018 = gendergrouping2018['Male'] /gendergrouping2018['Female']

gendergrouping2019 = df2019.groupby('Gender').count()
gendergrouping2019 = gendergrouping2019['num']

ratio2019 = gendergrouping2019['Male'] /gendergrouping2019['Female']

ratiodata = [['2017', ratio2017], ['2018', ratio2018], ['2019', ratio2019]]

# Create the pandas DataFrame
genderratiodf = pd.DataFrame(ratiodata, columns=['Year', 'Ratio'])

genderratioplot = genderratiodf.pivot_table(
    index =["Year"], values='Ratio',
).plot.bar(figsize=(10, 5))

genderratioplot.set_xlabel('Year')
genderratioplot.set_ylabel('Ratio Men:Women')
```

```python
#the ratio of women to men in each age group

agesratio = datasci.pivot_table(
    index =["Age"], columns='Gender', values='num', aggfunc='count'
)

agesratio = agesratio['Male']/agesratio['Female']


agesratioplot = agesratio.plot.bar(figsize=(10, 5))

agesratioplot.set_xlabel('Age')
agesratioplot.set_ylabel('Ratio Men:Women')
```

```python
#the ratio of women to men in each country
countryratio = datasci.pivot_table(
    index =["Residence"], columns='Gender', values='num', aggfunc='count'
)

countryratio = countryratio['Male']/countryratio['Female']

countryratio_sorted = countryratio.sort_values(ascending=False)


countryratioplot = countryratio_sorted.plot.bar(figsize=(20, 5))

countryratioplot.set_xlabel('Residence')
countryratioplot.set_ylabel('Ratio Men:Women')

#need to do data cleaning with residence
# need to merge Viet Nam and Vietnam
#need to merge united kingdom
#need to merge united states
#need to merge hong kong
#neet to make sure value counts arent too low to make these stats
```
